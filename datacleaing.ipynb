```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('D:\DISHWISE\DS FILES\DATASET (1).csv')
df1 = pd.read_csv('D:\DISHWISE\DS FILES\seasonb.csv')
df2 = pd.read_csv('D:\DISHWISE\DS FILES\\tamil_nadu_dataset.csv')
```

```python
unique_commodities = df['Commodity'].unique()
print("Unique Commodities:", unique_commodities)
print("Number of Unique Commodities:", len(unique_commodities))
```

```python
data= pd.merge(df, df1, on='Commodity', how='inner')
data
```

```python
data['Arrival_Date'] = pd.to_datetime(data['Arrival_Date'])
data['Year'] = data['Arrival_Date'].dt.year
data['Month'] = data['Arrival_Date'].dt.month
monthly_avg = data.groupby(['Year', 'Month', 'District', 'Market', 'Commodity', 'Variety'])[['Min_Price', 'Max_Price', 'Modal_Price']].mean().reset_index()
merged_data = pd.merge(df1, monthly_avg, on=['Commodity'], how='inner')
merged_data
```

```python
sorted_data = merged_data.sort_values(['Month', 'District', 'Market', 'Commodity', 'Variety'])
sorted_data = sorted_data.reset_index(drop=True)
print(sorted_data)
sorted_data
```

```python


commodity_variety_dict = sorted_data.groupby('Commodity')['Variety'].apply(lambda x: sorted(x.unique())).to_dict()

print(commodity_variety_dict)
```

```python
# prompt: map the Month with the name of the month according to number

month_map = {
    1: 'January',
    2: 'February',
    3: 'March',
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'October',
    11: 'November',
    12: 'December'
}
sorted_data['Month_Name'] = sorted_data['Month'].map(month_map)
print(sorted_data)
pd.drop(sorted_data['Month'])
sorted_data
```

```python
sorted_data.to_csv('FINAL.csv', encoding = 'utf-8-sig')
sorted_data
```

```python
import pandas as pd
import os

# Define the path to your CSV file and output directory
input_file_path = r'D:\DISHWISE\DS FILES\DATASET (1).csv'
output_dir = r'D:\DISHWISE\DS FILES\split_files'  # Adjust as necessary
os.makedirs(output_dir, exist_ok=True)

# Read the CSV file in chunks
chunk_size = 25000000  # 25 MB
chunk_number = 0

# Use chunksize to read in manageable chunks
for chunk in pd.read_csv(input_file_path, chunksize=chunk_size):
    # Create a new file for each chunk
    chunk_file_path = os.path.join(output_dir, f'dataset_part_{chunk_number}.csv')
    chunk.to_csv(chunk_file_path, index=False)
    print(f'Saved: {chunk_file_path}')
    chunk_number += 1

print("Splitting completed!")
```
